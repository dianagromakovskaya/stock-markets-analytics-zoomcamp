# Question 5. [Stocks] Dividend Yield
**Find the largest dividend yield for the same set of stocks**

Use the same list of companies (2222.SR,BRK-B, AAPL, MSFT, GOOG, JPM) and download all dividends paid in 2023. You can use get_actions() method or .dividends field in yfinance library (https://github.com/ranaroussi/yfinance?tab=readme-ov-file#quick-start)

Sum up all dividends paid in 2023 per company and divide each value by the closing price (Adj.Close) at the last trading day of the year.

Find the maximum value in % and round to 1 digit after the decimal point. (E.g., if you obtained \$1.25 dividends paid and the end year stock price is $100, the dividend yield is 1.25% -- and your answer should be equal to 1.3)



```python
for ticker in ('2222.SR', 'BRK-B', 'AAPL', 'MSFT', 'GOOG', 'JPM'):
  ticker_data = yf.Ticker(ticker)
  dividends = ticker_data.get_actions()
  dividends2023y = dividends[dividends.index.year == 2023].Dividends.sum()

  df_daily = yf.download(tickers=ticker,
                         start="2023-01-01",
                         end="2024-01-01",
                         interval="1d")
  last_day_closing_price = df_daily[df_daily.index == df_daily.index.max()]['Adj Close'][0]
  print(ticker)
  print(round(dividends2023y / last_day_closing_price * 100, 1))
```

## Sharpe Ratio

The Sharpe ratio is a metric used to assess the risk-adjusted performance of an investment. Calculated as the difference between the returns of the investment and the risk-free rate, divided by the investment’s standard deviation, the Sharpe ratio effectively communicates how much additional return an investor is receiving for the extra volatility endured for holding a riskier asset. Higher Sharpe ratios indicate more desirable investments.

```python
import math


annual_risk_free_rate = 0.02
daily_risk_free_rate = math.log(1 + annual_risk_free_rate) / 252

for ticker in ('2222.SR', 'BRK-B', 'AAPL', 'MSFT', 'GOOG', 'JPM'):
  df_daily = yf.download(tickers=ticker,
                         start="2023-01-01",
                         end="2024-01-01",
                         interval="1d")
  df_daily['Daily Return'] = np.log(df_daily['Adj Close'] / df_daily['Adj Close'].shift(1))
  excess_return = df_daily['Daily Return'] - daily_risk_free_rate
  sharpe_ratio = round(excess_return.mean() / df_daily['Daily Return'].std(), 2)
  print(f'\n Sharpe Ratio for {ticker}: {sharpe_ratio}')
```

## Beta

Beta is a measure of a stock's volatility in relation to the overall market. By definition, the market, such as the S&P 500 Index, has a beta of 1.0, and individual stocks are ranked according to how much they deviate from the market. A stock that swings more than the market over time has a beta above 1.0. If a stock moves less than the market, the stock's beta is less than 1.0.

```python
market_data = yf.download('^GSPC', start='2023-01-01', end='2024-01-01')
```

```python
market_data['Daily Return'] = market_data['Adj Close'].pct_change()
```

```python
for ticker in ('2222.SR', 'BRK-B', 'AAPL', 'MSFT', 'GOOG', 'JPM'):
  df_daily = yf.download(tickers=ticker,
                         start="2023-01-01",
                         end="2024-01-01",
                         interval="1d")
  df_daily['Daily Return'] = np.log(df_daily['Adj Close'] / df_daily['Adj Close'].shift(1))
  excess_return = df_daily['Adj Close'].pct_change()
  beta = round((df_daily['Daily Return'].cov(market_data['Daily Return'])) / market_data['Daily Return'].var(), 2)
  print(f'\n Beta for {ticker}: {beta}')
```

## Dividend Growth Rate

The dividend growth rate measures the year-over-year growth percentage of a company’s dividend, offering insight into the growth trend of a company’s payouts.

```python
# Average Dividend Growth Rate for the last 5 years
for ticker in ('2222.SR', 'BRK-B', 'AAPL', 'MSFT', 'GOOG', 'JPM'):
  ticker_data = yf.Ticker(ticker)
  divs = ticker_data.get_actions()
  divs_5y = divs[divs.index.year.isin(range(2019, 2024))].reindex()
  divs_5y = divs_5y.groupby(by=divs_5y.index.year).agg({'Dividends': 'sum'})
  divs_5y['growth_rate_yoy'] = divs_5y.Dividends / divs_5y.Dividends.shift(1) - 1
  avg_dgr_5y = round(divs_5y['growth_rate_yoy'].mean(), 2)
  print(f'Average Dividend Growth Rate for the last 5 years for {ticker} is: {avg_dgr_5y}')
```
